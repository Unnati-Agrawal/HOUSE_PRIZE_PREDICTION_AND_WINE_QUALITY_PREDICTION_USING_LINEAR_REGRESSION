```python
cols = ['residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'sulphates']

Q1 = data[cols].quantile(0.25)
Q3 = data[cols].quantile(0.75)
IQR = Q3 - Q1

new_data = data[~((data[cols] < (Q1 - 1.5 * IQR)) |(data[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
new_data.head()
```

```python
from sklearn.model_selection import train_test_split
features = new_data.drop(['quality','goodquality'], axis = 1)
labels = new_data['goodquality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = 0.4, random_state = 42) 
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size = 0.5, random_state = 42) 
print (len(labels), len(y_train), len(y_val), len(y_test))
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(X_train,y_train)
y_pred1 =  model1.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
score1=r2_score(y_test,y_pred1)
print('R2 score is: ',score1)
print('mean_sqrd_error(MAE) is: ',mean_squared_error(y_test,y_pred1))
print('root_mean_squared error(RMSE) is: ',np.sqrt(mean_squared_error(y_test,y_pred1)))
```

```python
model1val = LinearRegression()
model1val.fit(X_val,y_val)
y_pred1val =  model1val.predict(X_test)
score1val=r2_score(y_test,y_pred1val)
print('R2 score is: ',score1val)
print('mean_sqrd_error(MAE) is: ', mean_squared_error(y_test,y_pred1val))
print('root_mean_squared error(RMSE) is: ',np.sqrt(mean_squared_error(y_test,y_pred1val)))
```
